# Fill-in of matrix powers

```julia
using FundamentalsNumericalComputation
```

Here is the adjacency matrix of a graph representing a "small world" network featuring connections to neighbors and a small number of strangers.

```julia
A = load("smallworld.jld2")["A"]
graphplot(A,linealpha=0.5)
```

Because each node connects to relatively few others, the adjacency matrix is quite sparse.

```julia
spy(A,title="Nonzeros in A",color=:bluesreds)
```

The entries of $\mathbf{A}^k$ give the number of paths of length $k$ between pairs of nodes. (This would be "*k* degrees of separation" in a social network.) As $k$, grows, then, the density of $\mathbf{A}^k$ also grows.

```julia
plt = plot(layout=(1,3),legend=:none,size=(600,250))
for k in 2:4
    spy!(A^k,subplot=k-1,color=:bluesreds,title="A^$k")
end
plt
```
