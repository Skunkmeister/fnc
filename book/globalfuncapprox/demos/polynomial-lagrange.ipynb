# Lagrange polynomials

```julia
using FundamentalsNumericalComputation
```

In this chapter, the node and data vectors are most naturally indexed starting at zero. We use an `OffsetArray` to create a vector that works this way.

```julia
n = 5;
t = OffsetArray( [ 1, 1.5, 2, 2.25, 2.75, 3 ], 0:n )
```

Here is the definition of the cardinal Lagrange polynomial for $k=2$. First we define a polynomial $\phi$ that is zero at all the nodes except $i=k$. Then $\ell_2$ is found by normalizing $\phi$ by $\phi(t_k)$.

```julia
k = 2;
ϕ = x -> prod( x-t[i] for i in [0:k-1;k+1:n] );
ℓ_k = x -> ϕ(x) / ϕ(t[k]);
```

```julia
plot(ℓ_k,1,3)

y = OffsetArray(zeros(n+1),0:n)
y[k] = 1
scatter!(t,y,color=:black,
    xaxis=(L"x"),yaxis=(L"\ell_2(x)"),title="Lagrange cardinal function",legend=:none)
```

Observe that $\ell_k$ is _not_ between zero and one everywhere between the nodes.
