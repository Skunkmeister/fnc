# Polynomial interpolation error

```julia
using FundamentalsNumericalComputation
```

Consider the problem of interpolating $\log(x)$ at these nodes:

```julia
t = OffsetArray( [ 1, 1.6, 1.9, 2.7, 3 ], 0:4 );
```

Here $n=4$ and $f^{(5)}(\xi) = 4!/\xi^5$. For $\xi\in[1,3]$ we can say that $|f^{(5)}(\xi)| \le 4!$. Hence 

$$ |f(x)-p(x)| \le \frac{1}{5} \Phi(x).$$

```julia
Phi = x -> prod(x-ti for ti in t)
plot(x->0.2*Phi(x),1,3)
scatter!(t,zeros(size(t)),color=:black,
    xaxis=(L"x"), yaxis=(L"\Phi(x)"),title="Interpolation error bound",legend=:none)
```

The error is zero at the nodes, by the definition of interpolation. The error bound has one local extreme point between each consecutive pair of nodes.
