# Interpolation error for equispaced nodes

```julia
using FundamentalsNumericalComputation
```

We plot $|\Phi(x)|$ over the interval $[-1,1]$ with equispaced nodes for different values of $n$. 

```julia
plot(xaxis=(L"x"),yaxis=(:log10,L"|\Phi(x)|",[1e-25,1]),label="",legend=:bottomleft)

x = LinRange(-1,1,2501)
for n = 10:10:50
    t = LinRange(-1,1,n+1)
    Phi = [ prod(xk.-t) for xk in x ]
    scatter!(x,abs.(Phi),m=(1,stroke(0)),label="n=$n")
end
title!("Error indicator for equispaced nodes")
```

Each time $\Phi$ passes through zero at an interpolation node, the value on the log scale should go to $-\infty$, which explains the numerous cusps on the curves.

Two observations are important: First, the size of $|\Phi|$ decreases exponentially at each fixed location in the interval (because the spacing between curves is constant for constant increments of $n$). Second, $|\Phi|$ is larger at the ends of the interval than in the middle, by an exponentially growing factor.
