# Interpolation error for Chebyshev nodes

```julia
using FundamentalsNumericalComputation
```

Now we look at the error indicator function $\Phi$ for Chebyshev node sets.

```julia
plot(xaxis=(L"x"),yaxis=(:log10,L"|\Phi(x)|",[1e-18,1e-2]))
x = LinRange(-1,1,2501)
for n = 10:10:50
    theta = pi*(0:n)/n
    t = @. -cos(theta)                   
    Phi = [ prod(xk.-t) for xk in x ]
    plot!(x,abs.(Phi),m=(1,:o,stroke(0)),label="n=$n")
end
title!("Error indicator for Chebyshev nodes")
```

In contrast to the equispaced case, $|\Phi|$ decreases exponentially with $n$ uniformly across the interval.
